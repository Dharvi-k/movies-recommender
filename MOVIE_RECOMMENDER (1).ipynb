 First we need to convert these strings into actual Python lists using **ast.literal_eval** before proceeding because genre contains lists stored as strings.

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
def convert(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
df['cast']=df['cast'].apply(convert)
```

```python
def fetch_director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i["name"])
  return l
```

### Stemming with NLTK
While inspecting the 5,000 keywords from `cv.get_feature_names_out()`, I noticed many words with the same meaning but different forms (e.g., *class* / *classes*, *connection* / *connections*, *competition* / *compete*).  
To reduce redundancy and treat such variations as the same word, I used **NLTK**'s Porter Stemmer.  
Stemming helps by converting words to their root form, improving the quality of our feature set and making the model focus on meaning rather than grammatical variations.


```python
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d SamWorthington ZoeSaldana SigourneyWeaver JamesCameron')
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000, stop_words='english')
vectors=cv.fit_transform(df['tags']).toarray()
```

We first installed **NLTK** and imported the **PorterStemmer** to reduce words to their root form (e.g., "classes"  "class", "connections"  "connect").  
Stemming was applied to the entire `tags` column to normalize similar words.  
After stemming, we applied **CountVectorizer**, which builds a vocabulary of **unique tokens**.  
Since stemming merged variations into the same root, CountVectorizer only kept one form of each word and selected the top 5,000 unique tokens based on frequency.

Each movie is represented as a numerical vector, and we calculate the **cosine angle** between them instead of using Euclidean distance, because cosine works better for high-dimensional data.


```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[0]
```

**HOW OUR RECOMMEND FUNCTION WORK**

1. Find the index of the input movie (Avatar).

2. Extract the similarity row for that movie  gives a vector of similarity scores.

3. Sort in descending order  pick top 5 movies  these are your recommendations.

```python
def recommend(movie):
  movie_index=df[df['title']==movie].index[0]
  distance=similarity[movie_index]
  sorted_dist=distance.argsort()[::-1]
  top_5_indices=sorted_dist[1:6 ]
  for i in top_5_indices:
    print(df.iloc[i].title)
```

```python
recommend('Batman Begins')
```
